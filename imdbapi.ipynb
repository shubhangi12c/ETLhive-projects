```python
import requests
```

```python


from google.colab import userdata
userdata.get("api_key")


url = "https://imdb232.p.rapidapi.com/api/title/get-most-popular"

querystring = {"limit":"20","topMeterTitlesType":"ALL"}



headers = {
    "x-rapidapi-key":userdata.get("api_key"),
	"x-rapidapi-host": "imdb232.p.rapidapi.com"

}

response = requests.get(url, headers=headers, params=querystring)

#print(response.json())
data = response.json()

```

Now we want to extract data i.e 1. Movie name
2.Movie release year and
3.Rank of movies.
follow below steps

```python
#here we are going inside of nested dictionary one by one.
data.get("data",{})
```

```python
#here we are getting list which contains movie details in nested dictionary format.
data.get("data").get("topMeterTitles").get("edges",{})
```

```python
#we can check type
type(data.get("data").get("topMeterTitles").get("edges",{}))
```

```python
#if we want to see [0] position movie deatils.
data.get("data").get("topMeterTitles").get("edges",{})[0]
```

```python
#we will extract 1st movie details, here list contains a multiple dictionaries.
movie1 = data.get("data").get("topMeterTitles").get("edges")[0] # we are getting [0] index  details(dictionary details)
nd = movie1.get("node",{})                               #We are going inside of dict
nd
title = nd.get("titleText",{}).get("text",{})            #we are extracrting title of movie
release_year = nd.get("releaseYear",{}).get("year",{})   # we are extracting release year
rank = nd.get("meterRanking",{}).get("currentRank",{})   # we are extracting ranking
print(title,release_year,rank)                           #printing all details
```

```python
# now to get details of all movies , we will use for loop
nodes = data.get("data").get("topMeterTitles").get("edges") #storing list data in nodes.
for node in nodes:                                          #applying for loop
    nd = node.get("node",{})
    title = nd.get("titleText",{}).get("text",{})
    release_year = nd.get("releaseYear",{}).get("year",{})
    rank = nd.get("meterRanking",{}).get("currentRank",{})
    print(title,release_year,rank)
```

```python
#now we are going to save each movie details in one list
nodes = data.get("data").get("topMeterTitles").get("edges")
top_movie_data = [] #define empty list
for node in nodes:
    nd = node.get("node",{})
    title = nd.get("titleText",{}).get("text",{})
    release_year = nd.get("releaseYear",{}).get("year",{})
    rank = nd.get("meterRanking",{}).get("currentRank",{})
    #print(title,release_year,rank)
    top_movie_data.append([title,release_year,rank])
print(top_movie_data)
```

```python
# to save the extracted/collected data to a csv file
import csv
with open("Top_movies.csv","w",encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Movie Titles","Release Year","Rank"])
    writer.writerows(top_movie_data)
```

#now we will group all in a class so that we can create object in which we need to only pass api key and we can extract data easily

```python
class api_extracter:

    def __init__(self,api_key):
        self.api_key = api_key

    def get_data(self):
        url = "https://imdb232.p.rapidapi.com/api/title/get-most-popular"

        querystring = {"limit":"20","topMeterTitlesType":"ALL"}

        headers = {
            "x-rapidapi-key": userdata.get("api_key"),
            "x-rapidapi-host": "imdb232.p.rapidapi.com"
        }

        response = requests.get(url, headers=headers, params=querystring)

        #print(response.json())
        data = response.json()

        nodes = data.get("data").get("topMeterTitles").get("edges")
        top_movie_data = []
        for node in nodes:
            nd = node.get("node",{})
            title = nd.get("titleText",{}).get("text",{})
            release_year = nd.get("releaseYear",{}).get("year",{})
            rank = nd.get("meterRanking",{}).get("currentRank",{})
            #print(title,release_year,rank)
            top_movie_data.append([title,release_year,rank])
        return top_movie_data

    def save_files_csv(self,data):
        with open("Top_movies_2.csv","w",encoding="utf-8") as file:
            writer = csv.writer(file)
            writer.writerow(["Movie Titles","Release Year","Rank"])
            writer.writerows(data)
        print("File is saved succesfully")
```
